```python
# Fill missing values
def fill_missing(df):
    # Columns of missing numerical values and missing stings
    colmiss = [col for col in df.columns if df[col].isnull().any()]    
    colmiss_num = [col for col in colmiss if df[col].dtype==float or df[col].dtype==int]    
    colmiss_obj = list(set(colmiss)-set(colmiss_num))
    
    # Dealing with missing numeric values
    print("Missing numerical values:\n",df[colmiss_num].isnull().sum())    
    values = df[colmiss_num].median()
    df[colmiss_num] = df[colmiss_num].fillna(value=values)
    
    # Dealing with missing categorical values
    print("\nMissing categorical values:\n",df[colmiss_obj].isnull().sum())    
    # By foing through the dataset info, we see that missing categorical values 
    # actually refer to the given categorical item not being available
    df[colmiss_obj] = df[colmiss_obj].fillna(value='NotAvailable')
    
    return df
```

```python
df_train = fill_missing(df_train)
df_test = fill_missing(df_test)
```

```python
# Separating Categorical columns as ordinal or nominal    
cols = df_train.columns.tolist()
col_num = [col for col in cols if df_train[col].dtype=='float64' or \
           df_train[col].dtype=='int64']    
col_obj = list(set(cols)-set(col_num))

# We classify them by going through data description        
col_nom = ['MSZoning','Street','Alley','LotShape','LandContour','LotConfig',\
           'LandSlope','Neighborhood','Condition1','Condition2','BldgType',\
           'HouseStyle','RoofStyle','RoofMatl','Exterior1st','Exterior2nd',\
           'MasVnrType','Foundation','Heating','Electrical','GarageType',\
           'MiscFeature','SaleType','SaleCondition','Utilities','Fence']    
col_ord = [col for col in col_obj if col not in col_nom]

# Further classify nominal as having low or high cardinality
low_cardinality_nom = [col for col in col_nom if df_train[col].nunique()<10]
high_cardinality_nom = list(set(col_nom)-set(low_cardinality_nom))

# OneHotEncode nominal columns with cardinality less than 10 and rest will be
# label encoded
col_OHE = low_cardinality_nom
col_LE = col_ord+high_cardinality_nom

# One Hot encoding for nominal variables
OH_encoder = OneHotEncoder(handle_unknown='ignore', sparse=False)

df_train_OHE = pd.DataFrame(OH_encoder.fit_transform(df_train[col_OHE]))
df_test_OHE = pd.DataFrame(OH_encoder.transform(df_test[col_OHE]))
df_train_OHE.index = df_train.index
df_test_OHE.index = df_test.index
df_train_OHE.columns = OH_encoder.get_feature_names(df_train[col_OHE].columns.tolist())
df_test_OHE.columns = OH_encoder.get_feature_names(df_test[col_OHE].columns.tolist())

# Label Encoding for ordinal variables

label_encoder = LabelEncoder()
df_train_LE = df_train[col_LE].copy()
df_test_LE = df_test[col_LE].copy()
for col in col_LE:
    df_train_LE[col] = label_encoder.fit_transform(df_train[col])
    df_test_LE[col] = label_encoder.fit_transform(df_test[col])

# Combining all the columns and saving final dataset
df_train_num = df_train.drop(columns=col_obj,axis=1)
df_test_num = df_test.drop(columns=col_obj,axis=1)
df_train_final = pd.concat([df_train_num,df_train_OHE,df_train_LE],axis=1)
df_test_final = pd.concat([df_test_num,df_test_OHE,df_test_LE],axis=1)
```
